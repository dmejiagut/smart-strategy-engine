## 1) Subir y descomprimir el proyecto
Cuando te lo pida, selecciona el archivo **`Smart_Strategy_Engine.zip`**.

```python
import zipfile, os, glob
from google.colab import files

subidos = files.upload()  # elige el .zip del proyecto
zip_name = [n for n in subidos if n.lower().endswith('.zip')][0]
with zipfile.ZipFile(zip_name) as z:
    z.extractall('/content/proyecto')

# Detectar la carpeta que contiene app.py
app_path = glob.glob('/content/proyecto/**/app.py', recursive=True)[0]
PROY = os.path.dirname(app_path)
print('Proyecto en:', PROY)
os.chdir(PROY)
print('Archivos:', os.listdir(PROY))
```

## 3) Lanzar el dashboard y obtener el link
Espera ~15 segundos y abre el enlace **`https://...trycloudflare.com`** que aparezca abajo.

```python
import os, time, subprocess
os.chdir(PROY)

# Descargar cloudflared (túnel gratuito, sin registro)
!wget -q -nc https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64 -O /content/cloudflared
!chmod +x /content/cloudflared

# Lanzar Streamlit en segundo plano
subprocess.Popen(
    ['streamlit', 'run', 'app.py', '--server.port', '8501',
     '--server.headless', 'true', '--server.address', '0.0.0.0'],
    stdout=open('/content/streamlit_log.txt', 'w'),
    stderr=subprocess.STDOUT,
)
time.sleep(12)
print('Streamlit iniciado. Generando link público...\n')

# Abrir el túnel (el link aparece en la salida)
!/content/cloudflared tunnel --url http://localhost:8501 --no-autoupdate
```
